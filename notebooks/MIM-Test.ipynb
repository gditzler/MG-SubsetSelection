# Mutual Information and Metagenomic Data Analysis

Let a single taxonomic unit be viewed as a random variable, which is denoted as $X$. The RV $X$ has a support $\mathcal{X}$ and it is assumed throughout this work that it is a *discrete* RV. In feature subset selection we have an objective function $\mathcal{J}(X)$ that needs to be maximized. One popular scoring function, which is from information theory, is *mutual information*. Mutual information is made up of the more fundamental unit that is *entropy*. Entropy, from an information-theoretic perspective, is the measure of uncertainty in a RV. Mathematically, entropy is given by:   
$$
  H(X) = - \sum_{x\in\mathcal{X}} p_X(x) \log p_X(x)
$$
where $p_X(x)$ is the marginal distribution on the RV $X$. The mutual information between two RVs $X$ and $Y$ is given by:
$$
  I(X;Y) = H(X) - H(X|Y) = \sum_{y\in\mathcal{Y}}\sum_{y\in\mathcal{Y}} p_{XY}(x,y) \log\frac{p_{XY}(x,y)}{p_X(x)p_Y(y)}
$$
which is a measure of the variables' mutual dependence. Mutual information serves as a building block to implementing feature subset selection. In this notebook, we solely examine mutual information for determining feature importance.   

```python
import sys
sys.path.append("../src/")
import numpy as np 
import bmu
import utils
import matplotlib.pylab as plt
import mi

tag = "caporaso-gut"
biom_fp = "../data/caporaso-gut.biom"
map_fp = "../data/caporaso-gut.txt"
n_select = 25

data, samples, features = bmu.load_biom(biom_fp)
map_data = bmu.load_map(map_fp)
labels, label_map = utils.label_formatting(map_data, samples, "SEX", signed=True)
samples = np.array(samples)
features = np.array(features)
```

```python
mutual_info = mi.calc_mi(data=data, labels=labels)

plt.figure()
plt.plot(range(len(features)), np.sort(mutual_info)[::-1], lw=2)
plt.title("Mutual Information")
plt.xlabel("arbitrary index")
plt.ylabel("I(X;Y)")
a = axes([0.5, 0.5, .35, .35])
plt.plot(range(2000), np.sort(mutual_info)[::-1][:2000], lw=2)
plt.ylabel("I(X;Y)")

plt.figure()
plt.plot(range(len(features)), np.cumsum(np.sort(mutual_info)[::-1]), lw=2)
plt.title("CDF Mutual Information")
plt.xlabel("arbitrary index")
plt.ylabel("I(X;Y)")
a = axes([0.5, 0.4, .35, .35])
plt.plot(range(2000), np.cumsum(np.sort(mutual_info)[::-1][:2000]), lw=2)
```

```python
n_averages = 20
mutual_infos = np.zeros((len(features),))
plt.figure()
for n in range(n_averages):
    idx = np.random.randint(0, len(samples), len(samples))
    data_n = data[idx]
    labels_n = labels[idx]
    mutual_info = mi.calc_mi(data=data_n, labels=labels_n)
    if n == 0:
        idx_sorted = np.argsort(mutual_info)
    plt.plot(range(2000), mutual_info[idx_sorted][::-1][:2000], color='.7')
    mutual_infos += mutual_info

plt.plot(range(2000), mutual_infos[idx_sorted][::-1][:2000]/n_averages, color='0.')
plt.xlabel("arbitrary index")
plt.ylabel("I(X;Y)")
```
