The formatting of the results from `timerz.sh` are not in the prettiest of formats, so we can do a bit of work with sed to clean up the timing results and write a cleaned up set of text files that contains the evaluation times for JMI, MIM and QIIME. After the next cell we can plot the results in Python. 

```python
%%bash
ls ../outputs/timerz/jmi-*.time \
  | xargs cat \
  | grep real \
  | sed -e "s/real//g" \
  | tr '\t' ' ' \
  | sed -e "s/ //g" -e "s/m/ /g" -e "s/s//g" -e "s/^/60\*/g" -e "s/ /\+/g" \
  | bc -l > ../outputs/timerz/all-jmi.txt

ls ../outputs/timerz/mim-*.time \
  | xargs cat \
  | grep real \
  | sed -e "s/real//g" \
  | tr '\t' ' ' \
  | sed -e "s/ //g" -e "s/m/ /g" -e "s/s//g" -e "s/^/60\*/g" -e "s/ /\+/g" \
  | bc -l > ../outputs/timerz/all-mim.txt

ls ../outputs/timerz/cmim-*.time \
  | xargs cat \
  | grep real \
  | sed -e "s/real//g" \
  | tr '\t' ' ' \
  | sed -e "s/ //g" -e "s/m/ /g" -e "s/s//g" -e "s/^/60\*/g" -e "s/ /\+/g" \
  | bc -l > ../outputs/timerz/all-cmim.txt

ls ../outputs/timerz/mrmr-*.time \
  | xargs cat \
  | grep real \
  | sed -e "s/real//g" \
  | tr '\t' ' ' \
  | sed -e "s/ //g" -e "s/m/ /g" -e "s/s//g" -e "s/^/60\*/g" -e "s/ /\+/g" \
  | bc -l > ../outputs/timerz/all-mrmr.txt

ls ../outputs/timerz/qiime.time \
  | xargs cat \
  | grep real \
  | sed -e "s/real//g" \
  | tr '\t' ' ' \
  | sed -e "s/ //g" -e "s/m/ /g" -e "s/s//g" -e "s/^/60\*/g" -e "s/ /\+/g" \
  | bc -l > ../outputs/timerz/qiime.txt

ls ../outputs/timerz/lasso.time \
  | xargs cat \
  | grep real \
  | sed -e "s/real//g" \
  | tr '\t' ' ' \
  | sed -e "s/ //g" -e "s/m/ /g" -e "s/s//g" -e "s/^/60\*/g" -e "s/ /\+/g" \
  | bc -l > ../outputs/timerz/lasso2.txt
```

```python
X = np.array(range(100,1000,100))
mim_times = np.array([float(x) for x in open("../outputs/timerz/all-mim.txt","U").read()[:-1].split("\n")])
jmi_times = np.array([float(x) for x in open("../outputs/timerz/all-jmi.txt","U").read()[:-1].split("\n")])
cmim_times = np.array([float(x) for x in open("../outputs/timerz/all-cmim.txt","U").read()[:-1].split("\n")])
mrmr_times = np.array([float(x) for x in open("../outputs/timerz/all-mrmr.txt","U").read()[:-1].split("\n")])
qiime_times = np.ones(X.shape)*np.array([float(x) for x in open("../outputs/timerz/qiime.txt","U").read()[:-1].split("\n")])[0]
lasso_times = np.ones(X.shape)*np.array([float(x) for x in open("../outputs/timerz/lasso2.txt","U").read()[:-1].split("\n")])[0]

#plt.plot(X, jmi_times, label="JMI", c="b")
#plt.plot(X, mim_times, label="MIM", c="r")
#plt.plot(X, qiime_times, label="RF-QIIME", c="k")
plt.plot(X[1:], np.log(cmim_times)[1:], label="CMIM", c="g")
plt.plot(X[1:], np.log(mrmr_times)[1:], label="mRMR", c="c")
plt.plot(X[1:], np.log(jmi_times)[1:], label="JMI", c="b")
plt.plot(X[1:], np.log(mim_times)[1:], label="MIM", c="r")
plt.plot(X[1:], np.log(qiime_times)[1:], label="RF-QIIME", c="k")
plt.plot(X[1:], np.log(lasso_times)[1:], label="LASSO", c="y")
plt.xlabel('# of Features')
plt.ylabel('LOG(Runtime) (sec)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop={'size':22})
# NOTE THAT QIIME IS JUST PLOTTED AS A FUNCTION OF FEATURES FOR DEMO PURPOSES
```
